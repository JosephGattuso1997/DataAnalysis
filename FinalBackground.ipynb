```python
import matplotlib.pyplot as plt
import numpy as np

State, Mortality, Owner = [], [], []
with open('gun_violence_us.csv', 'r') as f:
    for item in f.read().split('\n')[1:-1]:
        Total_ = [n for n in item if n !='']
        record = item.split(',')
        State.append(record[0])
        State = State[-50:]
        Mortality.append(record[1])
        Mortality = Mortality[-50:]
        Owner.append(record[2])
        Owner = Owner[-50:]
```

```python
plt.scatter(State, Owner)
plt.rcParams["figure.figsize"] = (20,10)
plt.axis()
plt.xlabel('State')
plt.ylabel("Ownership")
plt.title('Percent Ownership by State')
plt.grid(True)                                                                                                   
plt.show()
```

```python

plt.bar(State, Mortality)
plt.rcParams["figure.figsize"] = (20,10)
plt.axis()
plt.xlabel('State')
plt.ylabel("Mortality by State per 100,000")
plt.title('Mortality per State')
plt.grid(True)                                                                                                   
plt.show()
```
